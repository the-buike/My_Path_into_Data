```python
#import necessary library
import pandas as pd
from IPython.display import HTML
import base64
```

```python
#Load the dataset
file_path = "DataProfessionalSurvey.xlsx"
dps_df = pd.read_excel(file_path, sheet_name="Data Professional Survey")
```

```python
dps_df.head(10)
```

```python
#show all columns
dps_df.columns
```

```python
#Drop unnecessary columns
columns_to_drop = [
    'Unique ID', 'Email', 'Date Taken (America/New_York)',
       'Time Taken (America/New_York)', 'Browser', 'OS', 'City', 'Country',
       'Referrer'

]
dps_df.drop(columns=columns_to_drop,inplace=True)
```

```python
#check general data sanity
dps_df.info()
```

```python
#check nulls
dps_df.isnull().sum()
```

```python
#fill missing or null values with Median (middle value in a sorted list)
# Select only numeric columns
numeric_cols = dps_df.select_dtypes(include='number').columns

# Fill missing values with median
for col in numeric_cols:
    median_val = dps_df[col].median()
    dps_df[col].fillna(median_val, inplace=True)
```

```python
#check nulls
dps_df.isnull().sum()
```

```python
# Fill missing education with the most frequent (mode) value
edu_col = 'Q12 - Highest Level of Education'
mode_val = dps_df[edu_col].mode()[0]
dps_df[edu_col].fillna(mode_val, inplace=True)
```

```python
#check nulls
dps_df.isnull().sum()
```

```python
#show all columns
dps_df.columns
```

```python
dps_df.head(10)
```

```python
#Remove the string in the "Current Yearly Salary (in USD)" columns
dps_df['Q3 - Current Yearly Salary (in USD)'] = dps_df['Q3 - Current Yearly Salary (in USD)'].str.replace('k', '', regex=False)
```

```python
dps_df.head(10)
```

```python
#find the avergae salary in "Current Yearly Salary (in USD)"
#1: Split the range into two columns
salary_split = dps_df['Q3 - Current Yearly Salary (in USD)'].str.split('-', expand=True)

#2: Convert both parts to numeric
salary_split[0] = pd.to_numeric(salary_split[0], errors='coerce')
salary_split[1] = pd.to_numeric(salary_split[1], errors='coerce')

#3: Calculate the average
dps_df['Average Salary (in USD)'] = (salary_split[0] + salary_split[1]) / 2
```

```python
dps_df.head(10)
```

```python
#clean the "Q4 - What Industry do you work in?" to only include others to generalise other careers that were not listed in the survey questions
dps_df['Q4 - What Industry do you work in?'] = dps_df['Q4 - What Industry do you work in?'].str.replace(r'^Other\s*\(Please Specify\):.*', 'Other', regex=True)
```

```python
dps_df.head(10)
```

```python
#clean the "Q1 - Which Title Best Fits your Current Role" to only include others to generalise other careers that were not listed in the survey questions
dps_df['Q1 - Which Title Best Fits your Current Role?'] = dps_df['Q1 - Which Title Best Fits your Current Role?'].str.replace(r'^Other\s*\(Please Specify\):.*', 'Other', regex=True)
```

```python
#clean the "Q1 - Which Title Best Fits your Current Role" to only include others to generalise other careers that were not listed in the survey questions
dps_df['Q8 - If you were to look for a new job today, what would be the most important thing to you?'] = dps_df['Q8 - If you were to look for a new job today, what would be the most important thing to you?'].str.replace(r'^Other\s*\(Please Specify\):.*', 'Other', regex=True)
```

```python
#clean the "Q11 - Which Country do you live in?" to only show the countries 
dps_df['Q11 - Which Country do you live in?'] = dps_df['Q11 - Which Country do you live in?'].str.extract(r'([^:,]+)$')
```

```python
#clean the "Q1 - Which Title Best Fits your Current Role" to only include others to generalise other careers that were not listed in the survey questions
dps_df['Q13 - Ethnicity'] = dps_df['Q13 - Ethnicity'].str.replace(r'^Other\s*\(Please Specify\):.*', 'Other', regex=True)
```

```python
#clean the "Q5 - Favorite Programming Language" to only include others to generalise other careers that were not listed in the survey questions
dps_df['Q5 - Favorite Programming Language'] = dps_df['Q5 - Favorite Programming Language'].str.replace(r'^Other:.*', 'Other', regex=True)
```

```python

#For asier Understanding i renamed the columns
dps_df = dps_df.rename(columns={
    'Time Spent': 'TimeSpent',
    'Q1 - Which Title Best Fits your Current Role?': 'JobTitle',
    'Q2 - Did you switch careers into Data?': 'CareerSwitch',
    'Q3 - Current Yearly Salary (in USD)': 'SalaryRange',
    'Q4 - What Industry do you work in?': 'Industry',
    'Q5 - Favorite Programming Language': 'FavoriteLanguage',
    'Q6 - How Happy are you in your Current Position with the following? (Salary)': 'Happiness_Salary',
    'Q6 - How Happy are you in your Current Position with the following? (Work/Life Balance)': 'Happiness_WorkLife',
    'Q6 - How Happy are you in your Current Position with the following? (Coworkers)': 'Happiness_Coworkers',
    'Q6 - How Happy are you in your Current Position with the following? (Management)': 'Happiness_Management',
    'Q6 - How Happy are you in your Current Position with the following? (Upward Mobility)': 'Happiness_Mobility',
    'Q6 - How Happy are you in your Current Position with the following? (Learning New Things)': 'Happiness_Learning',
    'Q7 - How difficult was it for you to break into Data?': 'DifficultyBreakingIn',
    'Q8 - If you were to look for a new job today, what would be the most important thing to you?': 'JobSearchPriority',
    'Q9 - Male/Female?': 'Gender',
    'Q10 - Current Age': 'Age',
    'Q11 - Which Country do you live in?': 'Country',
    'Q12 - Highest Level of Education': 'Education',
    'Q13 - Ethnicity': 'Ethnicity',
    'Average Salary (in USD)': 'Average_Salary'
})
```

```python
# Convert categorical fields
categorical_cols = [
    'JobTitle', 'CareerSwitch', 'Industry', 'FavoriteLanguage',
    'DifficultyBreakingIn', 'JobSearchPriority', 'Gender',
    'Country', 'Education', 'Ethnicity'
]
dps_df[categorical_cols] = dps_df[categorical_cols].astype('category')

#Convert TimeSpent to datetime or timedelt
dps_df['TimeSpent'] = pd.to_timedelta(dps_df['TimeSpent'], errors='coerce')
```

```python
#Remove (+) in the salaryrange
dps_df['SalaryRange'] = dps_df['SalaryRange'].str.replace('+', '', regex=False)



#Split SalaryRange
salary_split = dps_df['SalaryRange'].str.split('-', expand=True)

# Step 3: Convert both parts to numeric
salary_split[0] = pd.to_numeric(salary_split[0], errors='coerce')
salary_split[1] = pd.to_numeric(salary_split[1], errors='coerce')

#For rows 379 and 454, fill upper bound (if NaN) with the lower bound
for idx in [379, 454]:
    if pd.isna(salary_split.loc[idx, 1]):
        salary_split.loc[idx, 1] = salary_split.loc[idx, 0]

#Calculate the average and assign directly to df
dps_df.loc[[379, 454], 'Average_Salary'] = (
    salary_split.loc[[379, 454], 0] + salary_split.loc[[379, 454], 1]
) / 2
```

```python
#Download the cleaned data (70% cleaned) 
def create_download_link(dps_df, filename="cleanedsurvey.csv"):
    csv = dps_df.to_csv(index=False)
    b64 = base64.b64encode(csv.encode()).decode()
    return HTML(f'<a download="{filename}" href="data:text/csv;base64,{b64}">Download CSV</a>')

create_download_link(dps_df)
```
